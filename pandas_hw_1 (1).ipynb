```python
import pandas as pd
technologies = {
    'Courses':["Spark","PySpark","Python","pandas"],
    'Fee' :[20000,25000,22000,30000],
    'Duration':['30days','40days','35days','50days'],
              }
index_labels=['r1','r2','r3','r4']
df1 = pd.DataFrame(technologies,index=index_labels)

technologies2 = {
    'Course':["Spark","Java","Python","Go"],
    'Discount':[2000,2300,1200,2000]
              }
index_labels2=['r1','r6','r3','r5']
df2 = pd.DataFrame(technologies2,index=index_labels2)


print(df2), print(df1)
```

```python
#لدمج البيانات بناء على عمود .join()
# index

dfjoin=df1.join(df2, lsuffix="_left", rsuffix="_right", how='left')
dfjoin
```

```python
dfmerge = pd.merge(df1, df2, left_index=True, right_index=True, how='left')
print(dfmerge)
```

```python
# pandas merge - inner join by Column
dfmerge2 =pd.merge(df1,df2, how= "left")
dfmerge2
```

```python

# Merge DataFrames by Column
dfmerge3=pd.merge(df1,df2, on='Courses')
dfmerge3
```

```python
# When column names are different
dfcolnames=pd.merge(df1,df2, left_on='Courses', right_on='Course')
dfcolnames
```

```python
true_p = 5000
false_p = 2000
false_n = 1000
true_n = 2000
```

```python
#انشئ دالة تحسب الدقة وتقوم بطباعة النتيجة
def precision(TP,FP):
  #اكمل
  prec = TP / (TP+FP)
  print("precision score is : ", prec )
```

```python
#ثم استخدم الدالة بادخال البيانات بها> سواء بشكل مباشر او باسم المتغير
precision(true_p, false_p)
```

```python
#انشئ دالة تحسب الاسترجاع وتقوم بطباعة النتيجة
def recall(TP,FN):
  rec= TP / (TP+FN)
  print(f"recall score is: {rec}")
#اكمل
```

```python
recall(5000, 1000)
```

```python
def f1_score(recll, preci):
  x = (preci*recll)/(preci+recll)
  F1 = 2*x
  print("the F1 score is", F1)
  return F1
```

```python
f1_score(0.83 ,0.71)
```
